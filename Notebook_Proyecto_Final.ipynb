# Definición Formal

El autómata está definido por:
<center>
$E = (Q, \Sigma ,\delta, F, q_0)$
</center>
<br>
siendo $Q$ el conjunto de estados, $\Sigma$ es el alfabeto, $\delta$ es la función de transición, $F$ es el conjunto de estados de aceptación y $q_0$ el estado inicial.

<center>
### Conjunto de Estados

<center>
$Q = \{q_0,q_1,q_2,q_3,q_4,q_5,q_6,q_7,q_8,q_9,q_{10},q_{11},q_{12}\}$
</center>

<center>
### Estado inicial: 

<center>
${q_0}$ -> Aquí es dónde empieza el usuario a digitar sus respuestas en la encuesta.

<center>
### Alfabeto

<center>
$\Sigma=\{a, b, c, s, n, e, i, 1, 2, 3, 4, 5\}$ -> Son todas las posibles entradas.

<center>
### Conjunto de Estados de Aceptación
<center>
{$q_{12}$} -> Dónde debe terminar el autómata tras finalizar la encuesta, dado el caso que el crédito sea aprobado.

#### Implementación usando automata-lib 

A continuación, se presenta la implementación del diagrama de estados del autómata, utilizando la librería automata-lib.

```python
from automatalib.fa.dfa import DFA
dfa1 = DFA(
    states={'q0', 'q1', 'q2', 'q3', 'q4', 'q5', 'q6', 'q7', 'q8', 'q9', 'q10', 'q11', 'q12'},
    input_symbols={'s', 'n', '1', '2', '3', '4', '5', 'a', 'b', 'c', 'e', 'i'},
    transitions={
        'q0': {
            's': 'q1',
            'n': 'q2'
        },
        'q1': {
            'n': 'q3',
            's': 'q2'
        },
        'q2': {},
        'q3': {
            's': 'q4',
            'n': 'q2'
        },
        'q4': {
            '1': 'q5',
            '2': 'q5',
            '3': 'q5',
            '4': 'q5',
            '5': 'q5'
        },
        'q5': {
            's': 'q6',
            'n': 'q6'
        },
        'q6': {
            'n': 'q7'
        },
        'q7': {
            'a': 'q8',
            'b': 'q8',
            'c': 'q8'
        },
        'q8': {
            '1': 'q9',
            '2': 'q9',
            '3': 'q9',
            '4': 'q9'
        },
        'q9': {
            '1': 'q10',
            '2': 'q10',
            '3': 'q10',
            '4': 'q10',
            '5': 'q10'
        },
        'q10': {
            'e': 'q11',
            'i': 'q11'
        },
        'q11': {
            'n': 'q2',
            's': 'q12'
        },
        'q12': {}
    },
    initial_state='q0',
    final_states={'q12'}
)
```

# Funciones Adicionales
A continuación, se presentan unas funciones adicionales que permiten categorizar al cliente.
La funcion cambiarIngresos, realiza el cambio de una variable global ingresos dependiendo de la respuesta del usuario.

La función cambiarPrestamo, realiza el cambio de una variable glogal prestamo, dependiendo del tipo de crédito y la cantidad de dinero que necesita el usuario, dependiendo de las respuestas de él.

La función verificarFactibilidad, realiza un último filtro dónde se ejecuta un modelo matemático, dependiendo de las respuestas del usuario. Tiene como parámetros el tipo de crédito, el plazo en meses y si es empleado o independiente.
Primero, se verifica que la cuota mensual a pagar, sea menor a:
Sí el usuario es empleado, la cuota debe ser menor al 20% de sus ingresos.
Sí el usuario es independiente, la cuota debe ser menor al 15% de sus ingresos.

El porcentaje para independiente es menor, pues es un cliente con mayor riesgo al no tener ingresos estables.
Esta función devuelve Verdadero o Falso, dependiendo de sí cumple la condición en verificarFactibilidad.
El último caracter es 's' sí le aprueban el crédito; 'n' si es negado. (Para que esto afecte, debe pasar todos los filtros de las preguntas anteriores)

```python
ingresos = 1
prestamo = 1
tiempo = 1
cuota = 1

def cambiarIngresos(num):
    global ingresos
    if num == '1':
        ingresos = 1500000
    elif num == '2':
        ingresos = 2500000
    elif num == '3':
        ingresos = 3500000
    elif num == '4':
        ingresos = 4500000
    elif num == '5':
        ingresos = 5500000
        
def cambiarPrestamo(tipo, num):
    global prestamo
    mult = 1
    if tipo == 'b':    # Es crédito de vehículo
        mult = 10
    elif tipo == 'c':     # Es crédito de vivienda
        mult = 100
        
    if num == '1':
        prestamo = 1500000*mult
    elif num == '2':
        prestamo = 2500000*mult
    elif num == '3':
        prestamo = 3500000*mult
    elif num == '4':
        prestamo = 4500000*mult

        
def verificarFactibilidad(tipo, plazo, labor):
    global tiempo    # Tiempo es el plazo para pagar el préstamo en meses.
    global cuota
    
    if tipo == 'a':   # Es crédito de consumo
        if plazo == '1':
            tiempo = 6
        elif plazo == '2':
            tiempo = 12
        elif plazo == '3':
            tiempo = 24
        elif plazo == '4':
            tiempo = 36
        elif plazo == '2':
            tiempo = 48
        
        cuota = prestamo/tiempo
        
    elif tipo == 'b':    # Es crédito de vehículo
        if plazo == '1':
            tiempo = 12
        elif plazo == '2':
            tiempo = 24
        elif plazo == '3':
            tiempo = 48
        elif plazo == '4':
            tiempo = 72
        elif plazo == '2':
            tiempo = 84
            
        cuota = prestamo/tiempo
        
    elif tipo == 'c':     # Es crédito de vivienda
        if plazo == '1':
            tiempo = 48
        elif plazo == '2':
            tiempo = 96
        elif plazo == '3':
            tiempo = 120
        elif plazo == '4':
            tiempo = 144
        elif plazo == '2':
            tiempo = 180
            
        cuota = prestamo/tiempo
    
    if labor == 'e':     # Empleado
        if cuota < (ingresos*0.2):     # Si el valor de la cuota es menor al 20%, le da el crédito.
            return True
        else:
            return False
        
    elif labor == 'i':      #Independiente
        if cuota < (ingresos*0.15):     # Si el valor de la cuota es menor al 15%, le da el crédito.
            return True
        else:
            return False
```

# Implementación de la Encuesta

Finalmente, acá se realiza la implementación de la encuesta, donde al usuario se le realizan preguntas con sus posibles respuestas.
Aquí es dónde se llaman las funciones adicionales.

Después de realizada la encuesta, el usuario recibe una respuesta de sí fue aprobado o no.

```python
print("Bienvenido.")
print("A continuación, se le harán unas preguntas relacionadas con su análisis crediticio.")
print("NOTA: cualquier caracter que sea una letra, escribirlo en minúscula.")
print("- - - - - - - - - - - - - - - - -")

print("¿Es usted mayor de edad? ( s / n )")
respuesta1 = str(input())

print("¿Se encuentra reportado en centrales de riesgo? ( s / n )")
respuesta2 = str(input())

print("¿Ha iniciado su vida crediticia? ( s / n )")
respuesta3 = str(input())

print("¿De cuántos son sus ingresos? Ingrese el número de acuerdo a su rango de ingresos.")
print("     1. De $1.000.000 a $1.999.999")
print("     2. De $2.000.000 a $2.999.999")
print("     3. De $3.000.000 a $3.999.999")
print("     4. De $4.000.000 a $4.999.999")
print("     5. Más de $5.000.000")
respuesta4 = str(input())

#Aquí se cambia el valor de la variable ingresos, según lo que ingrese el usuario.
cambiarIngresos(respuesta4)

print("¿Tiene deudas? ( s / n )")
respuesta5 = str(input())

respuesta6 = 'n'
if respuesta5 == 's':
    print("¿Sus deudas superan el 50% de sus ingresos? ( s / n )")
    respuesta6 = str(input())
    
print("¿Qué tipo de préstamo necesita? Ingrese la letra de acuerdo a su interés.")
print("     a. Crédito de Consumo.")
print("     b. Crédito de Vehículo")
print("     c. Crédito de Vivienda")
respuesta7 = str(input())

print("¿Cuánto dinero necesita?")
if respuesta7 == 'a':
    print("     1. De $1.000.000 a $1.999.999")
    print("     2. De $2.000.000 a $2.999.999")
    print("     3. De $3.000.000 a $3.999.999")
    print("     4. De $4.000.000 a $5.000.000")
elif respuesta7 == 'b':
    print("     1. De $10.000.000 a $19.999.999")
    print("     2. De $20.000.000 a $29.999.999")
    print("     3. De $30.000.000 a $39.999.999")
    print("     4. De $40.000.000 a $50.000.000")
elif respuesta7 == 'c':
    print("     1. De $100.000.000 a $199.999.999")
    print("     2. De $200.000.000 a $299.999.999")
    print("     3. De $300.000.000 a $399.999.999")
    print("     4. De $400.000.000 a $500.000.000")
respuesta8 = str(input())

#Aquí se cambia el valor de la variable prestamo, según el tipo de crédito y lo que seleccione el usuario.
cambiarPrestamo(respuesta7, respuesta8)

print("¿En cuánto tiempo planea saldar el préstamo?")
if respuesta7 == 'a':
    print("     1. 6 meses.")
    print("     2. 12 meses.")
    print("     3. 24 meses.")
    print("     4. 36 meses.")
    print("     5. 48 meses.")
elif respuesta7 == 'b':
    print("     1. 12 meses.")
    print("     2. 24 meses.")
    print("     3. 48 meses.")
    print("     4. 72 meses.")
    print("     5. 84 meses.")
elif respuesta7 == 'c':
    print("     1. 4 años.")
    print("     2. 8 años.")
    print("     3. 10 años.")
    print("     4. 12 años.")
    print("     5. 15 años.")
respuesta9 = str(input())

print("¿Es empleado o independiente?")
print("     e. Empleado")
print("     i. Independiente")
respuesta10 = str(input())

#Aquí se hace un último filtro, dónde se hace un análisis matemático de las variables ingreso, prestamo y, dependiendo del resultado, otorgar o no el crédito.
verificar = verificarFactibilidad(respuesta7, respuesta9, respuesta10)

respuesta11 = 0
if verificar == True:
    respuesta11 = 's'
else:
    respuesta11 = 'n'

respuestaTotal = respuesta1 + respuesta2 + respuesta3 + respuesta4 + respuesta5 + respuesta6 + respuesta7 + respuesta8 + respuesta9 + respuesta10 + respuesta11
print("- - - - - - - - - - - - - - - - -")

print(respuestaTotal)
print("- - - - - - - - - - - - - - - - -") 

try:
    dfa1.validate_input(respuestaTotal)
    print("¡Felicitaciones! Tienes tu crédito aprobado con un valor de $", prestamo, " para pagar en un plazo máximo de ", tiempo, " meses.")
    print("Tus cuotas mensuales tienen el valor de: $", cuota)
except Exception as e:
    print("No fue posible otorgarle el crédito o hay un error en digitación.")
```
